```python
from datetime import datetime, timedelta
from pandas_datareader import data as pdr

import time 
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt


class Stock: 
    def __init__(self, stock_name, start_dt, end_dt):
    yf.pdr_override() # <== that's all it takes :-)

    self.stock_name = stock_name 
    self.start_dt = start_dt
    self.end_dt = end_dt 
    print("Name:{}, start_dt:{}, end_dt:{}".format(self.stock_name, self.start_dt, self.end_dt))
    self.data = pdr.get_data_yahoo(self.stock_name, start=self.start_dt, end=self.end_dt)

    self.ticker = yf.Ticker(self.stock_name)

    try: 
      self.beta = self.ticker.info['beta']
    except: 
      self.beta = 'N/A'

    self.timeseries = self.data['Close']
    self.dates = self.timeseries.index.to_frame()
    # self.one_year_return = self.get_annual_return(1)
    # self.two_year_return = self.get_annual_return(2)
    # self.five_year_return = self.get_annual_return(5)

    def get_annual_return(self, num_years):
    period = 365 *num_years
    print("Period:{}".format(period))
    try: 
      end = max(self.timeseries.index)
      print("end:{}".format(end))
      start = end - timedelta(days=period)

    except: 
      pass 

    try: 
      start = self.dates[self.dates['Date']<=start].iloc[-1].Date
      print("start:{}".format(start))

      ending_balance = self.timeseries.loc[end]
      beginning_balance = self.timeseries.loc[start]  

      return ((ending_balance - beginning_balance)/beginning_balance)/(period//365)

    except:
      print("\n Start date:{} is not available!".format(start))

    def get_monthly_return(self, num_months):
    period = 30 *num_months
    print("Period:{}".format(period))
    end = max(self.timeseries.index)
    print("end:{}".format(end))

    start = end - timedelta(days=period)
    try: 
      start = self.dates[self.dates['Date']<=start].iloc[-1].Date
      print("start:{}".format(start))

      ending_balance = self.timeseries.loc[end]
      beginning_balance = self.timeseries.loc[start]  

      return ((ending_balance - beginning_balance)/beginning_balance)/(period//30)

    except:
      print("\n Start date:{} is not available!".format(start))


    def get_monthly_std(self, num_months): 
    period = 30*num_months
    print("Period:{}".format(period))
    end = max(self.timeseries.index)
    print("end:{}".format(end))

    start = end - timedelta(days=period)
    try: 

      std_dev = self.data.loc[start:end]['Close'].std()

      print("start:{}".format(start))

      return std_dev

    except:
      print("\n Start date:{} is not available!".format(start))



    def plot(self):
    self.timeseries.plot()
    plt.title(self.stock_name)
```

```python
def get_risk_return_in_months(list_of_tickers, num_months): 
  mth_returns = []
  mth_stddevs = []

  for ticker in list_of_tickers: 
    ticker_yahoo = None
    while ticker_yahoo is None:
      try:
          # connect
          print(ticker)
          time.sleep(1)
          ticker_yahoo = Stock(ticker, start_dt, end_dt)
          mth_returns.append(ticker_yahoo.get_monthly_return(num_months))
          mth_stddevs.append(ticker_yahoo.get_monthly_std(num_months))
      except:
        mth_returns.append(0)
        mth_stddevs.append(1000)
        pass
  return mth_returns, mth_stddevs
```

```python
# import wikipedia
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize

import inflect

import spacy
nlp = spacy.load("en")
```

```python
def remove_commas(s):
  return s.replace("," , "")

def string_to_int(s):
  return int(s)

def remove_brackets(s):
    return re.sub("[\(\[].*?[\)\]]", "", s)

def clean_string(s): 
  return s.replace('\n', '').replace('\xa0', '').replace(",", "")

def concatenate_list_data(list):
    result= ''
    for element in list:
        result += " " 
        result += str(element)
    return result


def remove_stop_words(mystr): 
  text = mystr
  text_tokens = word_tokenize(text)

  tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]

  return tokens_without_sw

def get_unique_words(input_list):
  return list(set(input_list))


from nltk.corpus import words

def words_to_lowercase(input_list):   
  output_list= [] 
  for i in range(0, len(input_list)): 
    if plural_to_singular_word(input_list[i].lower()) in words.words():
      output_list.append(plural_to_singular_word(input_list[i].lower()))
    else: 
      output_list.append(plural_to_singular_word(input_list[i]))
  return output_list

def plural_to_singular_word(word): 
  p = inflect.engine()
  if p.singular_noun(word): 
    return p.singular_noun(word)
  else: 
    return word


def remove_non_english_words(mylist):
  english_words = []
  non_english_words = []
  for i in range(0, len(mylist)): 
    if mylist[i] in words.words(): 
      #print("{} is an english word".format(mylist[i]))
      english_words.append(mylist[i])
    elif mylist[i].lower() in words.words(): 
      #print("{} is an english word".format(mylist[i]))
      english_words.append(mylist[i])
    else: 
      #print("{} is NOT an english word".format(mylist[i]))
      non_english_words.append(mylist[i])
  
  non_english_words = plural_to_singular(non_english_words)

  mylist = non_english_words
  for i in range(0, len(mylist)): 
    if mylist[i] in words.words(): 
      #print("{} is an english word".format(mylist[i]))
      english_words.append(mylist[i])
    elif mylist[i].lower() in words.words(): 
      #print("{} is an english word".format(mylist[i]))
      english_words.append(mylist[i])
    else: 
      #print("{} is NOT an english word".format(mylist[i]))
      non_english_words.append(mylist[i])
  return english_words

def plural_to_singular(words): 
  singular_words = []
  p = inflect.engine()
  for i in range(0, len(words)):
    if p.singular_noun(words[i]): 
      #print("{} is singular ".format(words[i]))
      singular_words.append(p.singular_noun(words[i]))
    else: 
      #print("{} is not singular ".format(words[i]))
      singular_words.append(words[i])
  return singular_words


def get_word_similarity(word1, word2): 
  word1= nlp(word1)
  word2 = nlp(word2)
  return word1.similarity(word2)

def get_similarity_to_museum(input_list): 
  similarities=[]
  for i in range(0, len(input_list)):
    similarities.append((input_list[i], get_word_similarity('museum', input_list[i])))
  return similarities


def get_top_n_features(mystr, n): 
  input_list = remove_stop_words(mystr)
  input_list = get_unique_words(input_list)
  input_list = words_to_lowercase(input_list)
  english_words_only = remove_non_english_words(input_list)
  english_words_only = get_unique_words(english_words_only)
  similarities = get_similarity_to_museum(english_words_only)
  top_n_features = similarities_df.sort_values(by=['similarity'], ascending = False).head(n)
  top_n_features =  list(top_n_features['word'])  

  return top_n_features 


def get_relevant_features(mystr): 
  input_list = remove_stop_words(mystr)
  input_list = get_unique_words(input_list)
  input_list = words_to_lowercase(input_list)
  english_words_only = remove_non_english_words(input_list)
  english_words_only = get_unique_words(english_words_only)
  similarities = get_similarity_to_museum(english_words_only)
  similarities_df = pd.DataFrame(similarities, columns=['word', 'similarity'])
  mean = similarities_df['similarity'].mean()
  similarities_df = similarities_df[similarities_df['similarity']> mean]
  relevant_features =  list(similarities_df['word'])  

  return relevant_features 
```
